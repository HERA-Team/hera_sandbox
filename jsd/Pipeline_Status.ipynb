```python
config = ConfigParser(interpolation=ExtendedInterpolation())
config.read(config_file)
workflow = get_config_entry(config, 'WorkFlow', 'actions')
```

```python
def elapsed_time(log_lines):
    try:
        start = datetime.strptime(log_lines[0], '%a %b %d %H:%M:%S %Z %Y ')
    except:
        start = None
    try:
        end = datetime.strptime(log_lines[-1], '%a %b %d %H:%M:%S %Z %Y ')
    except:
        end = None
    
    if (end == None):
        return -1
    elif (start == None):
        return -2
    else:
        return ((end - start).seconds + 24.0*60*60*(end - start).days)/60.0
```

```python
def inspect_log_files(log_files, done_files):
    error_warned = False
    runtimes = []
    for log_file in log_files:
        with open(log_file,'r') as f:
            log_lines = f.readlines()
        runtimes.append(elapsed_time(log_lines))
        
        #If it completed running but there's no .out file
        if (runtimes[-1] > 0) and (log_file.replace('.log','.out') not in done_files) and ~error_warned:
            print '\n\nError Suspected in', log_file
            print '------------------------------------------------\n'
            print '\n'.join(log_lines)
            print '------------------------------------------------\n\n'
            error_warned = True
        
    runtimes = np.array(runtimes)
    finished_runtimes = runtimes[runtimes>1e-3]    
    average_runtime = np.mean(finished_runtimes)
    return average_runtime, np.sum(runtimes==-1)
```

```python
print '---------------\nPIPELINE REPORT\n---------------\n'

for job in workflow:
    scheduled = glob.glob(os.path.join(working_dir,'wrapper_*.' + job + '.*sh'))
    logged = glob.glob(os.path.join(working_dir,'*.' + job + '.*log'))
    done = glob.glob(os.path.join(working_dir,'*.' + job + '.*out'))
    average_runtime, nRunning = inspect_log_files(logged, done)
    
    print job + ':'
    print 'Average (non-zero) runtime:', average_runtime, 'minutes' 
    print len(scheduled), '\t|\t', len(done), '\t|\t', nRunning
    print 'sched\t|\tdone\t|\trunning\n'
    
    #print len(scheduled), '\t', len(logged), '\t', len(done), '\n'
```
