```python
from hera_cal import io
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pyuvdata import UVData, UVCal
```

```python
basefile = '/lustre/aoc/projects/hera/H1C_IDR2/2458099/2458099/'
date = 'zen.2458099.'
#jds = ['37158','36413','35667']
#jds = ['37158','36413','35667','34921']
jds = ['22245','22991','23737']
for jd in jds:
    uvd = UVData()
    uvd.read_miriad(basefile + date + jd + '.xx.HH.uvOR')
    uvd.select(ant_str='36x_36x')
    uvd.history += 'Antenna 36x autocorrelations extrated for unit testing by Josh Dillon on March 2, 2018.' 
    uvd.write_miriad(date + jd + '.xx.HH.uvOR_36x_auto_only', clobber=True)
    cal = UVCal()
    cal.read_calfits(basefile + date + jd + '.HH.uv.omni.calfits')
    cal.select(antenna_nums=36, jones=-5)
    cal.history += 'Antenna 36x extrated for unit testing by Josh Dillon on March 2, 2018.'
    cal.write_calfits(date + jd + '.HH.uv.omni.calfits_36x_only', clobber=True)
```
